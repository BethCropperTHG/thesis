```python
def gaus(x,amp,mu,sigma, r,beta):
    return( amp * (1 - r/100) * np.exp((-(x - mu)**2)/(2 * sigma**2)))
    
def skew(x,amp,mu,sigma,r,beta):   
    return(amp * (r/100) * np.exp((x-mu)/beta) * erfc( (x-mu)/(sigma * np.sqrt(2))  + sigma/(beta*np.sqrt(2))))
```

```python
fig = plt.figure(figsize = [6,6])
ax = fig.add_subplot(111)

amp = 100
mu = 75
sigma = 5
r = 20
beta = 11

x = np.linspace(0,100,1001)

ygaus = gaus(x,amp,mu,sigma, r,beta)
yskew = skew(x,amp,mu,sigma, r,beta)
ygf3 = gf3([amp,mu,sigma, r,beta],x)

ax.set_ylim(0,100)
ax.set_xlim(0,100)

ax.plot(x,ygaus, 'xkcd:dried blood', label = 'Gaussian Component', lw = 2.5)
ax.plot(x,yskew, 'xkcd:shit brown', label = 'Exponential Component', lw = 2.5)
ax.plot(x,ygf3, 'xkcd:puke', label = 'Total', lw = 2.5)

ax.legend()

plt.savefig("peak_split.png")
```
