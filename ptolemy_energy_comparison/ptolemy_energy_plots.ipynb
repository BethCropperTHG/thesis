```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
from cycler import cycler
%matplotlib inline
```

# Plot (d,p) , (p,d) comparison at different energies

## get (p,d) data

```python
pd_folders = np.sort(glob.glob('data/p,d/*'))#[[1,0,2]]
pd_folders
```

```python
pd_data = []

for folder in pd_folders:
    energy_files = np.array(glob.glob(folder + '/output_files/*'))
    jplist = []
    for file in energy_files:
        jplist.append(file[129:133])
    energy_files = energy_files[np.argsort(jplist)]
    dists = []
    for f in energy_files:
        df = np.loadtxt(f)
        dists.append(df)
    pd_data.append(dists)
    
```

## get (d,p) data

```python
dp_folders = np.sort(glob.glob('data/d,p/*'))#[[1,0,2]]
dp_folders
```

```python
dp_data = []

for folder in dp_folders:
    energy_files = np.array(glob.glob(folder + '/output_files/*'))
    jplist = []
    for file in energy_files:
        jplist.append(file[129:133])
    energy_files = energy_files[np.argsort(jplist)]
    dists = []
    for f in energy_files:
        df = np.loadtxt(f)
        dists.append(df)
    dp_data.append(dists)
```

## set plot rc parameters

```python
colours = ['blue','green','red','cyan','orange','forestgreen','plum','magenta','lime','purple','cyan',\
           'pink','darkgreen','rosybrown','maroon',\
          'magenta','olive','peru','yellow','darkblue',\
           'fuchsia','crimson']

plt.rc('text',usetex=True)
plt.rc('font', size=18)
plt.rc('font', weight='bold')
plt.rc('font', family='serif')          # controls default text sizes
plt.rc('axes', titlesize=15)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
plt.rc('legend', fontsize=12)    # legend fontsize
plt.rc('figure', titlesize=15)  # fontsize of the figure title
plt.rc('ytick', direction='in')
plt.rc('xtick', direction='in')

plt.rc('xtick.major',size=7)
plt.rc('ytick.major',size=7)
plt.rc('xtick.minor',size=6)
plt.rc('ytick.minor',size=6)

plt.rc('axes', xmargin = 0)
plt.rc('axes', ymargin = 0)

plt.rc('axes', prop_cycle = cycler('color', ['black','orange','red','xkcd:light brown','xkcd:green','blue','purple']))
```

## make plot

```python
fig = plt.figure(figsize = [10,7])


ax11 = fig.add_subplot(231)
ax21 = fig.add_subplot(232)
ax31 = fig.add_subplot(233)
ax12 = fig.add_subplot(234)
ax22 = fig.add_subplot(235)
ax32 = fig.add_subplot(236)



for l in pd_data[0]:
    ax11.plot(l[:,0],l[:,1])
for l in pd_data[1]:
    ax21.plot(l[:,0],l[:,1])
for l in pd_data[2]:
    ax31.plot(l[:,0],l[:,1])
    
for l in dp_data[0]:
    ax12.plot(l[:,0],l[:,1])
for l in dp_data[1]:
    ax22.plot(l[:,0],l[:,1])
for l in dp_data[2]:
    ax32.plot(l[:,0],l[:,1])
    
ax11.set_ylabel('Cross Section (mb/sr)')    
ax12.set_ylabel('Cross Section (mb/sr)')    
ax12.set_xlabel('Angle(degrees)')   
ax22.set_xlabel('Angle(degrees)')   
ax32.set_xlabel('Angle(degrees)')

ax11.set_title('$(p,d)$, Lab Energy = 15 MeV')
ax21.set_title('$(p,d)$, Lab Energy = 22 MeV')
ax31.set_title('$(p,d)$, Lab Energy = 44 MeV')
ax12.set_title('$(d,p)$, Lab Energy = 8 MeV')
ax22.set_title('$(d,p)$, Lab Energy = 15 MeV')
ax32.set_title('$(d,p)$, Lab Energy = 30 MeV')

handles = ('$\\ell = 0$','$\\ell = 1$','$\\ell = 2$','$\\ell = 3$','$\\ell = 4$','$\\ell = 5$','$\\ell = 6$')
ax31.legend(handles)

fig.tight_layout()

fig.savefig('d,p_p,d_momentum__energy_matching.png')
```
